```python
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import librosa
import matplotlib.pyplot as plt

# Load YAMNet model from TF Hub
model = hub.load('https://tfhub.dev/google/yamnet/1')
```

```python
# Upload a sample .wav file or use your own
from IPython.display import Audio
import os

# Load and resample audio
sample_file = 'audio.wav'  # replace with your own file
waveform, sample_rate = librosa.load(sample_file, sr=16000)
print("Sample rate:", sample_rate)
Audio(sample_file)
```

```python
# Run prediction
scores, embeddings, spectrogram = model(waveform)
scores_np = scores.numpy().mean(axis=0)

# Load class names
class_map_path = tf.keras.utils.get_file('yamnet_class_map.csv',
    'https://raw.githubusercontent.com/tensorflow/models/master/research/audioset/yamnet/yamnet_class_map.csv')

import pandas as pd
class_map_df = pd.read_csv(class_map_path)
class_names = class_map_df['display_name'].to_numpy()

# Show top 5 predictions
top_n = 5
top_indices = np.argsort(scores_np)[::-1][:top_n]

for i in top_indices:
    print(f"{class_names[i]}: {scores_np[i]:.3f}")
```
